# Llama2 模型测试

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# 模型路径：改成你本地的路径
model_path = "/home/NCUT/teacher/xmy/MRAG/PMI/hugging_cache/Llama-2-7b-hf"

# 加载分词器和模型
tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForCausalLM.from_pretrained(model_path, torch_dtype=torch.float16, device_map="cuda:3")
```

```python
# prompt 的设计
# prompt = '''
# <s>[INST] <<SYS>>
# You are a helpful assistant.
# <</SYS>>

# Context:
# The Eiffel Tower is located in Paris and was built in 1889.

# Question:
# When was the Eiffel Tower built? [/INST]
'''
```

```python
# 准备输入
prompt = "who is the president of USA?"
inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

# 生成文本
outputs = model.generate(
    **inputs,
    max_new_tokens=200,
    do_sample=True,
    top_k=50,
    top_p=0.95,
    temperature=0.7
)

# 解码输出
response = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(response)
```
