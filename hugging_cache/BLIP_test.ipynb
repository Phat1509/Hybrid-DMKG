# BLIP模型的测试

(使用环境是blip2opt)

```python
from transformers import Blip2Processor, Blip2ForConditionalGeneration
from PIL import Image
import torch

device = torch.device("cuda:3")

# 加载模型
processor = Blip2Processor.from_pretrained("/home/NCUT/teacher/xmy/MRAG/PMI/hugging_cache/BLIP2-OPT")
model = Blip2ForConditionalGeneration.from_pretrained(
    "/home/NCUT/teacher/xmy/MRAG/PMI/hugging_cache/BLIP2-OPT",
    torch_dtype=torch.float16,
    device_map={"": device}
)

# 加载图像
image = Image.open("/home/NCUT/teacher/xmy/MRAG/VLKEB-main/datasets/VLKEB_images/mmkb_images/m.0_7z2/bing_2.jpg").convert("RGB")
text = "Question: What is in the image? Answer:"

# 预处理输入
inputs = processor(images=image, text=text, return_tensors="pt")

# 设置 tensor 到设备，并保持类型
inputs["pixel_values"] = inputs["pixel_values"].to(device, dtype=torch.float16)
inputs["input_ids"] = inputs["input_ids"].to(device)
if "attention_mask" in inputs:
    inputs["attention_mask"] = inputs["attention_mask"].to(device)

# 生成回答
outputs = model.generate(
    **inputs,
    max_new_tokens=100,
    num_beams=5,
    no_repeat_ngram_size=3
)

# 解码输出
print(processor.decode(outputs[0], skip_special_tokens=True))
```

```python
# 加载图像
image = Image.open("/home/NCUT/teacher/xmy/MRAG/VLKEB-main/datasets/VLKEB_images/mmkb_images/m.0_9wr/bing_18.jpg").convert("RGB")
text = "Qustion:What is in the image? Answer: "

# 预处理输入
inputs = processor(images=image, text=text, return_tensors="pt")

# 设置 tensor 到设备，并保持类型
inputs["pixel_values"] = inputs["pixel_values"].to(device, dtype=torch.float16)
inputs["input_ids"] = inputs["input_ids"].to(device)
if "attention_mask" in inputs:
    inputs["attention_mask"] = inputs["attention_mask"].to(device)

# 生成回答
outputs = model.generate(
    **inputs,
    max_new_tokens=100,
    num_beams=5,
    no_repeat_ngram_size=3
)

# 解码输出
text_decode = processor.decode(outputs[0], skip_special_tokens=True)
```

```python
print(text_decode)
```

```python
def extract_answer(text: str, keyword: str = "Answer:") -> str:
    if keyword in text:
        return text.split(keyword, 1)[1].strip()
    else:
        return text.strip()

```

```python
answer = extract_answer(text_decode)
print(answer)
```
